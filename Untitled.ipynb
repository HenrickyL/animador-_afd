```python
dot = Digraph('The Round Table',format='png')
dot  #doctest: +ELLIPSIS
dot.node('A', 'King Arthur',)
dot.node('B', 'Sir Bedevere the Wise', color='blue',shape='')
dot.node('L', 'Sir Lancelot the Brave')
dot.edges(['AB', 'AL'])
dot.edge('B', 'L', constraint='false')
print(dot.source)
dot
```

```python
dot.render('test-output/round-table.gv', view=True) 
```

```python
dot = Digraph()
colors = ['green','red']
dot.attr('node', shape='point')
dot.node("")
dot.attr('node', shape='circle')
dot.node("A")
dot.node("B", color='red')


dot.clear()
#d.attr('edge', shape='circle')
dot.edge('','A',color='white')
dot.edge('A','B',color='green')
dot.edge('B','C',color=colors[1])
dot.attr(label=r'\n\nEntity \textbf{Relation} Diagram\ndrawn by NEATO')
dot.attr(fontsize='20')
dot.clear()
dot.attr(label=r'zzzz')
dot.attr(fontsize='15')
dot.view()
```
